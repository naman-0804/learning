```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA 
```

```python
data={
    'Age':[25, 30, 35, 40, 45,55],
    'Income':[50000, 60000, 80000, 90000, 120000,150000],
    'Spending':[2000, 3000, 4000, 5000, 7000,9000],
    'Savings':[10000, 15000, 20000, 25000, 30000,40000]
}
df=pd.DataFrame(data)
```

```python
scaler=StandardScaler()
scaled_data=scaler.fit_transform(df)
pca=PCA(n_components=2)
pca_data=pca.fit_transform(scaled_data)
pca_df=pd.DataFrame(data=pca_data, columns=['PC1', 'PC2'])
explained_variance=pca.explained_variance_ratio_
print("Explained Variance Ratios:", explained_variance)
print(np.round(explained_variance*100,2))
```

```python
plt.figure(figsize=(8,6))
plt.scatter(pca_df['PC1'], pca_df['PC2'], c='blue', edgecolor='k', s=100)
plt.title('PCA of Customer Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
print(pca_df)
```
